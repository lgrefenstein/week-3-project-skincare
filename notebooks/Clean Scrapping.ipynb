```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

results = []

for products_page in range(1, 14):
    url = f"https://www.lookfantastic.com/c/health-beauty/organic-natural/view-all-natural-organic/?facetFilters=en_beauty_formulation_content%3AClean%7CcategoryLevel2%3ASkin+Care&pageNumber={products_page}"
    response = requests.get(url)
    soup = BeautifulSoup(response.content, "html.parser")

    clean_products = soup.find_all("a", class_="product-item")

    for c in clean_products:

        name_product = c.get("data-title")
        link_product = "https://www.lookfantastic.com" + c.get("href")

        product_details = requests.get(link_product)
        detailed_soup = BeautifulSoup(product_details.content, "html.parser")

        try:
            brand_product = detailed_soup.find("a", class_="override-pdp-brand").get_text(strip=True)
        except AttributeError:
            continue

        try:
            ingredients_product = detailed_soup.find("div", attrs={"aria-labelledby": "ingredients"}).find("p").get_text(strip=True)
        except AttributeError:
            continue

        results.append({"Product Name": name_product,
                        "Product Brand": brand_product,
                        "Ingredients": ingredients_product,
                        "Clean Product": True,

        })

df = pd.DataFrame(results)
df.to_csv('clean_skincare_v1.csv', index=False)

df
```
