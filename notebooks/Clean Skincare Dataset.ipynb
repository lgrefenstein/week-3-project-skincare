```python
import pandas as pd

df = pd.read_csv("clean_skincare_v1.csv")

df.isna().sum()

df = df.drop_duplicates()

df.columns = df.columns.str.lower().str.replace(' ', '_')

df['ingredients'] = (
    df['ingredients']
    .str.replace('.', ',', regex=False)
    .str.replace('*', '', regex=False)
    .str.replace('\xa0', ' ', regex=False)
)

df['product_name'] = (df['product_name'].str.replace('\xa0', ' ', regex=False))
```

```python
uncategorized = []

category_keywords = {
    'Serum': ['serum', 'essence', 'ampoule'],
    'Moisturizer': ['moisturizer', 'moisturiser', 'cream', 'lotion', 'hydrating', 'lait', 'crème', 'balm'],
    'Sunscreen': ['sunscreen'],
    'Treatment/Active': ['retinol', 'vitamin', 'peptide', 'niacinamide', 'glycolic', 'salicylic', 'brightening', 'exfoliating', 'scrub', 'pigment', 'rescue', 'peel', 'patch'],
    'Cleanser': ['face cleanser', 'face wash', 'cleansing gel', 'foam cleanser', 'facial cleanser', 'cleanser', 'cleansing', 'micellar']
 }

def categorize_product(product_name):

    product_name = product_name.lower()

    matches = []

    for category, keywords in category_keywords.items():
        for keyword in keywords:
            if keyword in product_name:
                matches.append(category)
                break

    if matches:
        return ', '.join(matches)
    else:
        return 'Uncategorized'
```

```python
df['product_category'] = df['product_name'].apply(categorize_product)

df = df[df['product_category'] != 'Uncategorized']
df = df[df['product_category'] != 'Face Oil']
df = df[df['product_category'] != 'Toner']
df = df[df['product_category'] != 'Face Mist']
df = df[df['product_category'] != 'Lip Treatment']
df = df[df['product_category'] != 'Eye Treatment']
df = df[df['product_category'] != 'Sunscreen']



df["product_name"] = (
    df["product_name"]
    .str.replace(r"\(?\b\d+(\.\d+)?\s?(ml|g|kg|oz|l)\b\)?|\(Worth\b[^)]+\)\s*$|\(?\b\d+\s?Pack\b\)?|\bRefill\b|\(?\b\d+\s?Pairs\b\)?|\b\d+\s?Dose(s)?\b|\(?\bVarious\s+(Sizes|Shades)\b\)?|\bBundle\b|\bREFILL\b|\bJAR\b", "", regex=True)
    .str.replace(r"\s+", " ", regex=True)
    .str.strip()
)

df = df.drop(index=[337, 45, 287, 360, 99, 326, 250, 170, 143, 159, 107, 64, 251, 32])

df = df.drop_duplicates()

df
```
