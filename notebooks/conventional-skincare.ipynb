# Conventional Skincare Dataset Preparation

## Introduction

In this notebook, we prepare a conventional skincare dataset for comparison with the clean skincare dataset.

Our goal is to:
- load the conventional products dataset
- keep only relevant skincare products
- identify major conventional brands
- select a balanced sample of products
- export the cleaned dataset in the same structure as the clean dataset

```python
import pandas as pd
```

## Data Collection

We load the conventional products dataset and inspect its structure.
At this stage, we only want to understand the available columns and the overall size of the dataset.

```python
# Load the conventional products dataset
df = pd.read_csv("../data/conventional_products.csv")
```

```python
df.shape
```

```python
df.head()
```

```python
df.columns.tolist()
```

```python
# Select only the relevant columns for our analysis
df = df[["Brand", "Name", "Ingredients"]]

# Rename columns to match the structure of the clean dataset
df = df.rename(columns={
    "Brand": "Product Brand",
    "Name": "Product Name"
})
```

```python
df.head()
```

```python
# Combine brand and product name into one column
df["Product Name"] = df["Product Brand"] + " " + df["Product Name"]
```

```python
df.head()
```

```python
# Add column to indicate these are conventional products
df["Clean Product"] = False
```

```python
df.head()
```

```python
df.columns
```

```python
# Reload original dataset to access 'Label' again
df_full = pd.read_csv("../data/conventional_products.csv")
```

```python
df_full["Label"].value_counts()
```

```python
skincare_labels = [
    "Moisturizer",
    "Cleanser",
    "Face Mask",
    "Treatment",
    "Eye cream"
]
```

```python
# Filter dataset to only skincare products
df_filtered = df_full[df_full["Label"].isin(skincare_labels)]
```

```python
df_filtered["Label"].value_counts()
```

```python
df_filtered.shape
```

```python
# Count how many products each brand has
brand_counts = df_filtered["Brand"].value_counts()

brand_counts.head(20)
```

```python
selected_brands = [
    "SEPHORA COLLECTION",
    "CLINIQUE",
    "SHISEIDO",
    "MURAD",
    "PETER THOMAS ROTH",
    "KIEHL'S SINCE 1851",
    "DR. JART+",
    "LANCÔME",
    "LA MER",
    "ESTÉE LAUDER"
]
```

```python
# Keep only selected brands
df_brands = df_filtered[df_filtered["Brand"].isin(selected_brands)]
```

```python
df_brands["Brand"].value_counts()
df_brands.shape
```

```python
# Take up to 15 products per brand
df_sampled = df_brands.groupby("Brand").head(15)
```

```python
df_sampled.shape
df_sampled["Brand"].value_counts()
```

```python
df_sampled["Product Name"] = df_sampled["Brand"] + " " + df_sampled["Name"]
```

```python
df_final = df_sampled[["Brand", "Product Name", "Ingredients"]]
```

```python
df_final["Clean Product"] = False
```

```python
df_final.columns = ["Product Brand", "Product Name", "Ingredients", "Clean Product"]
```

```python
df_final.to_csv("conventional_skincare_final.csv", index=False)
```

```python
df_final.head()
```

```python
df_final.shape
```

## Data Cleaning

In this section, we perform basic cleaning steps to ensure consistency and usability of the dataset.

This includes:
- handling missing values
- standardizing text fields
- removing duplicates

```python
df_final.isna().sum()
```

```python
(df_final["Ingredients"].str.strip() == "").sum()
```

```python
df_final["Product Name"].duplicated().sum()
```

```python
df_final["Product Name"] = df_final["Product Name"].str.strip()
df_final["Product Brand"] = df_final["Product Brand"].str.strip()
df_final["Ingredients"] = df_final["Ingredients"].str.strip()
```
